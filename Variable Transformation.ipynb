### variable transformation : means trying to convert the column into normal distributed manner:

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import scipy.stats as stats 
```

```python
#load dataset
data=pd.read_csv("D:/Top Mentor/ml/New Folder/30 july/house_price.csv")
```

```python
#we apply variable transformation on LotArea
```

```python
#create a function for graph

def plotting(data, col):
    plt.figure(figsize=(15,3))
    
    plt.subplot(1,3,1)
    data[col].hist(bins=20)
    
    plt.subplot(1,3,2)
    data[col].plot(kind='kde',color='r')
    
    plt.subplot(1,3,3)
    stats.probplot(data[col],plot=plt)
    
    plt.show()
```

```python
plotting(data,'LotArea')
```

### log Transformation

```python
data['LotArea_log']=np.log(data['LotArea'])
```

```python
plotting(data,'LotArea_log')
```

### Reciprocal

```python
 
data['LotArea_reci'] = 1 / data['LotArea']
```

```python
plotting(data,'LotArea_reci')
```

### sqrt

```python
data['LotArea_srt'] = np.sqrt(data['LotArea'])
```

```python
plotting(data,'LotArea_srt')
```

###  exponentinal

```python
np.exp(4)
```

```python
data['LotArea_exp'] = data['LotArea'] ** (1/1.5)
```

```python
plotting(data,'LotArea_exp')
```

### box cox 

```python
data['LotArea_bx'] , lambda_ = stats.boxcox(data['LotArea'])
```

```python
plotting(data,'LotArea_bx')
```

### YEo-johnson Technique

```python
 data['LotArea_yeo'] , lambda_ = stats.yeojohnson(data['LotArea'])
plotting(data,'LotArea_yeo')
```

# Now we using all technique in GrLivArea column by using forloop

```python
data['GrLivArea']
```

```python
data['GrLivArea'].describe()
```

```python
for i in data.columns:
    if i == 'GrLivArea':
        print('original')
        plotting(data,i)  # original feature 
       
        print('using log')
        data['GrLivArea_log'] = np.log(data[i])
        plotting(data,'GrLivArea_log')  # 
        
        print('using reciprocal')
        data['GrLivArea_reci'] = 1 / data[i]
        plotting(data,'GrLivArea_reci')
        
        print('using sqrt')
        data['GrLivArea_sqrt'] = np.sqrt(data[i])
        plotting(data,'GrLivArea_sqrt')
        
        print('using exponential')
        data['GrLivArea_exp'] = data[i] ** (1/1.5)
        plotting(data,'GrLivArea_exp')
        
        print('using boxcox')
        data['GrLivArea_bx'],lambda_ = stats.boxcox(data[i])
        plotting(data,'GrLivArea_bx')
        
        print('using YEo-johnson Technique ')
        data['GrLivArea_yeo'],lambda_ = stats.yeojohnson(data[i])
        plotting(data,'GrLivArea_yeo')
        
        
```
